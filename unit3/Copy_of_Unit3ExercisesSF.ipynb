## The First Four Moments: Mean, Variance, Skew, and Kurtosis

When we talk about a probablity dist, we often care about at four things:
- The expected value, or where a distribution is centered (mean)
- The amount of variation from that centered value (variation)
- Whether or not our distribution favors, leans towards, or skews towards certain values. (skew)
- How often to expect unusually high or low values. (kurtosis)

###Playing with Discrete Distributions

**Task9**:

Plot the pdfs (techincally the pmfs) of the following distributions: $Bernoulli(p=0.7)$, $Binomial(n=10,p=0.7)$, $Poisson(λ=7)$, $Geometric(p=0.7)$, $HyperGeometric(N=50,k=35,n=15)$, $Categorical(p=[0.05,0.3,0.05,0.1,0.2,0.4])$

Additionally, compute the first four moments for all those distributions--except the Categorical distribution.

````python
bernoulli = pz.Bernoulli(0.7)
binomial = pz.Binomial(10, 0.7)
poisson = pz.Poisson(7)
geometric = pz.Geometric(0.7)
hyperGeometric = pz.HyperGeometric(50, 35, 15)
categorical = pz.Categorical(p=[0.05,0.3,0.05,0.1,0.2,0.4])

bernoulli.plot_pdf()
binomial.plot_pdf()
poisson.plot_pdf()
geometric.plot_pdf()
hyperGeometric.plot_pdf()
categorical.plot_pdf()

print("bernoulli")
print(bernoulli.moments())
print("binomial")
print(binomial.moments())
print("poisson")
print(poisson.moments())
print("geometric")
print(geometric.moments())
print("hyperGeometric")
print(hyperGeometric.moments())
````

**Task10**:

Describe what process each of the previous discrete distributions models, and their support.

Hint: Make sure to read any of the preliZ documentation, the wikipedia articles, or wolfram alpha.

Bernoulli: Supports x values of 1 and 0. It displays the probability of x = 1 happening or x = 0 happening based on the "p" argument, which encodes probability of the distribution.

Binomial: Supports x values from range 0 to n (argument defined in class declaration). Encodes the amount of times that either of two possibilities will happen give p (probability, also defined in class declaration).

Poisson:Poisson supports all x values that fit the description of a natural number. This distribution displays Models the number of events that occur within a set interval of time. The argument "mu" tells the class how many times a event will occur in a given interval.

Geometric: Geometric distributions support x values of all natural numbers greater than zero. Shows the probability that an event will occur in a "success" in a given number of bernoulli trials. P encodes probability and must be set upon class declaration.

HyperGeometric: HyperGeometri is similar to geometric in that it has to do with the chance of something happening in a given number of trials. However, it adds the additional parameters of n (how many trials), N (the population of possibilities with no replacement), and k, which is how many good outcomes there are in the population N. It has a range of x is equal to all numbers within the range such that [max(0, n-N+k), min(k,n)]
