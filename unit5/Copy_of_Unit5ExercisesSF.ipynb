**Task5**:

Create and run a stright line linear regression model on the howell data set with height predicting weight. Visualize your model and its posterior.

```python
with pm.Model() as nh_model:
    #priors
    α = pm.Normal("α", mu=0, sigma=3)
    β = pm.Normal("β", mu=0, sigma=10)

    mean = pm.Deterministic("mu", α + β * howell['height'])

    σ = pm.HalfNormal("σ", sigma=8)

    #likelihood
    #the linear part of the model is put directly into mu
    y = pm.Normal("y_pred",
                  mu=mean,
                  sigma=σ,
                  observed=howell["weight"])

    idata_nh_b = pm.sample()
```

```python
az.plot_posterior(idata_nh_b, var_names=["~mu"], figsize=(12, 3))
```

```python
#grabs 100 posterior samples
posterior = az.extract(idata_nh_b, num_samples=100)

# grabbing x values for graphing.
x_plot = xr.DataArray(
    np.linspace(howell["height"].min(), howell["height"].max(), 50),
    dims="plot_id"
    )

# this creates the expected line, the path we predict temperature and
#rented bike count deviates from
mean_line = posterior["α"].mean() + posterior["β"].mean() * x_plot

#These are a 50 lines that our model came up with
lines = posterior["α"] + posterior["β"] * x_plot

#set up plot
_, ax = plt.subplots()

#plots 50 other lines our model came up with.
ax.plot(x_plot, lines.T, c="C1", alpha=0.2, label="lines")

#plots the mean line
ax.plot(x_plot, mean_line, c="C0", label="mean line")

#plot the raw data
ax.plot(howell["height"], howell["weight"], "C2.", zorder=-3)

#label axes and create legend
ax.set_xlabel("height")
ax.set_ylabel("ght")
```

**Task6**:

Run a posterior predictive check for model_howell, plot the posterior predictive distribution, and comment on the fit of the model you made in Task5.

```python
pm.sample_posterior_predictive(idata_nh_b, model=nh_model,  extend_inferencedata=True)

az.plot_ppc(idata_nh_b, num_pp_samples=250, colors=["C1", "C0", "C1"])
```

While the tail of our model actually matches the second peak relatively well, overshooting just a little bit, its completely missing the mark on the first peak. In total I would say that this model is ok at what it does but its definitely not super useful and can be improved.

```python
#grabs the expected line, or line of best fit formula
mean_line = idata_nh_b.posterior["mu"].mean(("chain", "draw"))

#creates some x values to run through our line formula
height = np.random.normal(howell.height.values, 0.01)
idx = np.argsort(height)
x = np.linspace(height.min(), height.max(), 15)

#grabs the 94% HDI and 50% HDI, and sets them up for graphing.
y_pred_q = idata_nh_b.posterior_predictive["y_pred"].quantile(
    [0.03, 0.97, 0.25, 0.75], dim=["chain", "draw"]
)
y_hat_bounds = iter(
    [
        PchipInterpolator(height[idx], y_pred_q[i][idx])(x)
        for i in range(4)
    ]
)

#plots raw data and our line of best fit
_, ax = plt.subplots()
ax.plot(howell.height, howell.weight, "C2.", zorder=-3)
ax.plot(howell.height[idx], mean_line[idx], c="C0")


#graphs the 94% and 50% HDIs
for lb, ub in zip(y_hat_bounds, y_hat_bounds):
    ax.fill_between(x, lb, ub, color="C1", alpha=0.5)

#labels
ax.set_xlabel("height")
ax.set_ylabel("weight")
```

**Task7**:

Create and visualize the posterior of a generalized linear model that improves the on the fit of the model you made in Task5.

```python
with pm.Model() as model_exp:

    α = pm.Normal("α", mu=0, sigma=10)
    β = pm.Normal("β", mu=0, sigma=10)

    δ = pm.Normal("δ", mu=0, sigma=10)

    μ = pm.Deterministic("μ", pm.math.exp(α + β * howell.height))

    σ = pm.HalfNormal("σ", pm.math.exp(δ))

    y_pred = pm.Normal("y_pred", mu = μ, sigma = σ, observed = howell.weight)

    idata_exp = pm.sample()
```

```python
az.plot_posterior(idata_exp, var_names=["~μ"], figsize=(12, 3))
```

```python
pm.sample_posterior_predictive(idata_exp, model=model_exp,  extend_inferencedata=True)
```

**Task8**:

Run a posterior predictive check, and visualize the posterior predictive distribution, for the model you made in Task7.

Comment on the fit.

```python
#grabs the expected line, or line of best fit formula
mean_line = idata_exp.posterior["μ"].mean(("chain", "draw"))

#creates some x values to run through our line formula
height = np.random.normal(howell.height.values, 0.01)
idx = np.argsort(height)
x = np.linspace(height.min(), height.max(), 15)

#grabs the 94% HDI and 50% HDI, and sets them up for graphing.
y_pred_q = idata_exp.posterior_predictive["y_pred"].quantile(
    [0.03, 0.97, 0.25, 0.75], dim=["chain", "draw"]
)
y_hat_bounds = iter(
    [
        PchipInterpolator(height[idx], y_pred_q[i][idx])(x)
        for i in range(4)
    ]
)

#plots raw data and our line of best fit
_, ax = plt.subplots()
ax.plot(howell.height, howell.weight, "C2.", zorder=-3)
ax.plot(howell.height[idx], mean_line[idx], c="C0")


#graphs the 94% and 50% HDIs
for lb, ub in zip(y_hat_bounds, y_hat_bounds):
    ax.fill_between(x, lb, ub, color="C1", alpha=0.5)

#labels
ax.set_xlabel("height")
ax.set_ylabel("weight")
```

The fit is significantly better. It succesfully encapsulates both modes found within the observed data and while it does undershoot and overshoot a lot. The very tail end of this model lines up perfectly with observed data. Compared to the previous model, this represents a significant improvement and I think would be much more useful for predicting weight as a function of height.
